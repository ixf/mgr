```python
def seqc(seq):
    last = None
    count = 0
    counts = []
    for item in seq:
        if item != last:
            counts.append([last, count])
            count = 1
            last = item
        else:
            count += 1
        
    counts.append([last, count])
    return counts

q = DataFrame(seqc(list(df.filename)))
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding

def make_file_model(vocab_size):
    model = Sequential()
    model.add(Embedding(vocab_size, 16, input_length=max_length-1))
    model.add(LSTM(70))
    model.add(Dense(vocab_size, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
bdf = to_blocks(df)
```

```python
bdf
```

```python
t = Tokenizer(filters='')
filenames = list(set(bdf.filename))
txt = ' '.join(filenames)
t.fit_on_texts([txt])
encoded = t.texts_to_sequences([list(bdf.filename)])[0]

vocab_size = len(t.word_index) + 1

sequences = list()
sequence_length = 16
for i in range(sequence_length, len(encoded)):
    sequence = encoded[i-sequence_length:i+1]
    sequences.append(sequence)
max_length = max([len(seq) for seq in sequences])
sequences = pad_sequences(sequences, maxlen=max_length, padding='pre')
sequences = array(sequences)

X, y = sequences[:,:-1],sequences[:,-1]
y = to_categorical(y, num_classes=vocab_size)
```

```python
model = make_file_model(vocab_size)
model.fit(X, y, epochs=100, verbose=1)
```

```python
def generate_seq(model, tokenizer, max_length, seed_text, n_words):
    in_text = seed_text
    result = []
    for _ in range(n_words):
        encoded = tokenizer.texts_to_sequences([in_text])[0]
        encoded = pad_sequences([encoded], maxlen=max_length, padding='pre')
        print(encoded)
        yhat = model.predict_classes(encoded, verbose=0)
        out_word = ''
        for word, index in tokenizer.word_index.items():
            if index == yhat:
                out_word = word
                break
        result.append(out_word)
    return result
```

```python
some_input = list(bdf.filename[1:19])
print(generate_seq(model, t, max_length-1, some_input, 1))
print(bdf.filename[16:20])
```
